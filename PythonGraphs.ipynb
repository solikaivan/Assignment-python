```python
from matplotlib import pyplot as plt
import numpy as np
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the slope and intercept of the line
slope = 2
intercept = 1

# Define the x-axis values
x = range(0, 10)

# Calculate the y-axis values using the equation of a line
y = [slope * i + intercept for i in x]

# Plot the line
plt.plot(x, y)

# Set the title and axis labels
plt.title("Plot of a Straight Line")
plt.xlabel("x")
plt.ylabel("y")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of points in each line
n = 360

# Define the x-axis values for the first line
x1 = np.linspace(0, 10, n)

# Define the y-axis values for the first line using a sine wave
y1 = np.sin(x1)

# Define the x-axis values for the second line
x2 = np.linspace(0, 10, n) + np.pi/2

# Define the y-axis values for the second line using a cosine wave
y2 = np.cos(x2)

# Plot the first line
plt.plot(x1, y1, label="Line 1")

# Plot the second line
plt.plot(x2, y2, label="Line 2")

# Set the title and axis labels
plt.title("Plot of Two Intersecting Wavy Lines")
plt.xlabel("x")
plt.ylabel("y")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the angle values in degrees
angles = range(0, 361)

# Convert the angles to radians
radians = [angle * np.pi / 180 for angle in angles]

# Define the sine and cosine values
sine_values = [np.sin(angle) for angle in radians]
cosine_values = [np.cos(angle) for angle in radians]

# Plot the sine and cosine curves
plt.plot(angles, sine_values, label="Sine")
plt.plot(angles, cosine_values, label="Cosine")

# Set the title and axis labels
plt.title("Plot of Sine and Cosine Functions")
plt.xlabel("Angle (degrees)")
plt.ylabel("Value")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
